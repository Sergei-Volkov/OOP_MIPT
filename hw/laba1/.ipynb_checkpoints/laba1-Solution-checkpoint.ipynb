Отрицательные числа больше не используем, можете удалить их из списка.  
Преобразуйте ваш список с неотрицательными числами (или создайте новый) следующим образом: вместо числа i получите соответствующую букву из английского алфавита. Учтите, что нумерация начинается с 0. Т.е 0 соответствует 'a', 25 - 'z', 3 - "d"

Создайте файл strange_rna.txt и запишите в него ТОЛЬКО СОГЛАСНЫЕ буквы ("y" считаем согласной, всего согласных - 21) вашего списка в том же порядке, например


Считайте содержимое файла strange_rna.txt в строку strange_rna (именно из только что созданного файла, а не из списка!)

````python
#ваш код здесь
with open("strange_rna.txt") as file:
    strange_rna = file.read()
strange_rna
````

Напишите функцию, которая для двух аргументов - строки РНА и строки-триплета будет возвращать количество вхождений триплета в РНА. (например для РНА "bccbcbcb" количество вхождений триплета "cbc" равно 2ум)

````python
def num_triplets(rna, triplet):
    #ваш код здесь
    count = 0
    for i in range(len(rna)):
        if rna[i:].startswith(triplet):
            count = count + 1
    return count
````

Создайте словарь "интересных" триплетов triplets: для каждого триплета внесите в словарь его значение-свойство  
Триплеты с "интересными" свойствами:
"wnd" - восхитительность, "str" - странность, "btf" - красивость (не путать с восхитительностью!), "plt" - вежливость, "fnn" - веселость, "phd" - умность, "gvm" - государственность 

````python
triplets = {"wnd":"восхитительность", "str":"странность", "btf":"красивость", "plt":"вежливость", "fnn":"Веселость", "phd":"умность", 
           "gvm":"государственность"}
#ваш код здесь
````

Подсчитайте вхождения интересных триплетов в данный участок РНА. 
Создайте функцию, которая будет создавать файл отчета results.txt: для каждого описания триплета подпишите через двоеточие риск возникновения такого свойства в белке. Если вхождений триплета в участок нет, то вероятность равна 0.05, если вхождение одно - то 0.2, если два и более, то 0.6. Последней строкой добавляется строка вывода: "Повышенный риск мутаций в участке у свойств:" а далее через запятую перечисляются свойства с риском возникновения 0.1 и более, если таких свойств нет - то слово "нет"  
Например для rna="wnnwndphdppphhphd" отчетом будет  
"
восхитительность:0.2  
странность:0.05  
красивость:0.05  
вежливость:0.05  
веселость:0.05  
умность:0.6  
государственность:0.05  
Повышенный риск мутаций в участке у свойств:умность
"

````python
def make_result(rna, triplets, warn_lvl=0.1, file_name="results.txt"):
    #ваш код здесь
    with open(file_name, "w") as out:
        with_high_risk = []
        for key in triplets.keys():
            num_ins = num_triplets(rna, key)
            if num_ins == 0:
                probability = 0.05
            elif num_ins == 1:
                probability = 0.2
            else:
                probability = 0.6
                
            if probability >= warn_lvl:
                with_high_risk.append(key)
            
            out_string = triplets[key]+":"+str(probability)+"\n"
            out.write(out_string)
        outline = "Повышенный риск мутаций в участке у свойств:"
        outsumm = ",".join([triplets[key] for key in with_high_risk])
        if outsumm == "":
            outsumm = "нет"
        out.write(outline + outsumm)
            
````

````python
make_result(strange_rna, triplets)
strange_rna
````

````python
counts = {x:0 for x in triplets.keys()}
from tqdm import tqdm_notebook as tqdm
N = 10000
for i in tqdm(range(10000)):
    rand_nums = [randint(-5, 25) for x in range(2800)]
    new_list = list(map(get_letter, [x for x in rand_nums if x >= 0]))
    rnn = "".join([x for x in new_list if x not in vowels])
    for triplet in triplets.keys():
        num = num_triplets(rnn, triplet)
        counts[triplet] += num
````

````python
for triplet, proba in sorted(counts.items(), key=lambda item: item[1], reverse=True):
    print(f"{triplet} has probability {proba / N}")
````
